```python
import requests
import os
import json
import pandas as pd
```

```python
api_key = os.environ["4sq"]
```

```python
url= 'https://api.foursquare.com/v3/autocomplete?query=Restaurants&ll=46.87310%2C-96.78644&radius=4000'
```

```python
headers = {
    "Accept": "application/json",
    "Authorization": api_key
}
```

```python
response = requests.request("GET", url, params=params, headers=headers)

print(response.text)
```

```python
json_object = json.dumps(response.json(), indent=4)
 
with open("4Rsqsample.json", "w") as outfile:
    outfile.write(json_object)
```
